```python
#Check missing values
print("=====Missing Values Before Cleaning=====")
print(df.isnull().sum())
```

```python
#Example:Drop rows with too many missing values
df=df.dropna(how='all')
```

```python
#Example:Fill numerical columns with median
df=df.fillna(df.median(numeric_only=True))

```

```python
#Example:Fill categorical columns with mode
for column in df.select_dtypes(include=['object']).columns:
    df[column]=df[column].fillna(df[column].mode()[0])
```

```python
print("=====Missing Values After Cleaning=====")
print(df.isnull().sum())
```

```python
#2.Handle Duplicates
print("Duplicate rows before:",
df.duplicated().sum())
df=df.drop_duplicates()
print("Duplicate rows after:",
df.duplicated().sum())
```

```python
#Check columns and detect likelyconsumption/cost/year columns
print("All columns")
print(df.columns.tolist())
```

```python
# try to auto-detect likely column names
consumption_col = None
cost_col = None
year_col = None

for c in df.columns:
    lc = c.lower()
    if any(k in lc for k in ['consump', 'gallon', 'water']):
        if consumption_col is None: consumption_col = c
    if any(k in lc for k in ['cost', 'price', 'usd']):
        if cost_col is None: cost_col = c
    if 'year' in lc:
        if year_col is None: year_col = c

print("\nDetected columns (you can override these manually if needed):")
print("consumption_col ->", consumption_col)
print("cost_col        ->", cost_col)
print("year_col        ->", year_col)

# If detection failed, set them manually, e.g.
# consumption_col = 'ConsumptionGallons'
# cost_col = 'CostUSD'
# year_col = 'Year'
```

```python
#Records per year
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,4))
sns.countplot(x=year_col, data=df)
plt.title("Number of records per Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Interpretation of your above chart

```python
#Consumption and cost trend over time
# Step 5 â€” Trends Over Time
import matplotlib.pyplot as plt

# Group by year and take mean consumption & cost
trend = df.groupby(year_col)[[consumption_col, cost_col]].mean()

plt.figure(figsize=(10,5))
plt.plot(trend.index, trend[consumption_col], marker='o', label='Avg Consumption')
plt.plot(trend.index, trend[cost_col], marker='s', label='Avg Cost')
plt.title("Trends in Water Consumption & Cost Over Time")
plt.xlabel("Year")
plt.ylabel("Average Value")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Interpretation of your above chart

Chart Type:Line chart

This is a time trend line chart showing water consumption and cost across years.

Year(x-axis).

Avg Consumption(blue line,left y-axis).

Avg Cost (orange line,left-axis but very small values).

The chart shows that while average water consumption has fluctuated,it has generally declined from 2020 to 2024.On the other hand,average water costs remain stable and relatively low in comparision.This suggests that changes in consumption are not strongly reflected in the recorded costs,possibly due to subsidies, fixed pricing,or data scaling differences.

```python
#Correlation and scatter plot
# Step 6 â€” Relationship between Consumption and Cost

# 1. Correlation check
corr = df[[consumption_col, cost_col]].corr()
print("Correlation between consumption and cost:")
print(corr)

# 2. Scatter plot
plt.figure(figsize=(7,5))
plt.scatter(df[consumption_col], df[cost_col], alpha=0.5)
plt.title("Relationship between Water Consumption and Cost")
plt.xlabel("Water Consumption (Gallons)")
plt.ylabel("Cost (USD)")
plt.grid(True)
plt.tight_layout()
plt.show()

```

##Interpretation of Scatter plot
Chart type :Scatter plot.

This chart is showing the relationship between Water Consumption(Gallons) and Cost(USD)

Features used:

X-axis -->Water Consumption (Gallons)

Y-axis -->Cost (USD)

The scatter plot clearly demonstrates a strong linear relationship between water consumption and cost.Customer who consume more water consistently pay higher costs,confirming that consumption is the primary driver of water bills.

```python
# Group by year to see total and average consumption & cost
yearly_trends = df.groupby(year_col).agg({
    consumption_col: ['sum', 'mean'],
    cost_col: ['sum', 'mean']
}).reset_index()

# Rename columns for clarity
yearly_trends.columns = [year_col, 'Total_Consumption', 'Avg_Consumption',
                         'Total_Cost', 'Avg_Cost']

print(yearly_trends)

# Plot yearly total consumption
plt.figure(figsize=(8,5))
sns.lineplot(data=yearly_trends, x=year_col, y="Total_Consumption", marker="o")
plt.title("Yearly Trend of Total Water Consumption")
plt.xlabel("Year")
plt.ylabel("Total Consumption (Gallons)")
plt.tight_layout()
plt.show()

# Plot yearly average cost
plt.figure(figsize=(8,5))
sns.lineplot(data=yearly_trends, x=year_col, y="Avg_Cost", marker="o", color="red")
plt.title("Yearly Trend of Average Water Cost")
plt.xlabel("Year")
plt.ylabel("Average Cost (USD)")
plt.tight_layout()
plt.show()
```

##Interpretation of Yearly trend of water consumption and Yearly trend of water cost


These two charts together give both a macro view (total consumption across all users) and a micro view (average cost per customer).

By comparing them, we can answer questions like:

Does higher total consumption always lead to higher average costs?

Are costs rising due to price changes rather than consumption changes?

Is water demand steady, seasonal, or growing?

In short, the charts provide time-trend insights into how water consumption and costs evolve year by year â€” useful for forecasting, policy planning, and resource management.
